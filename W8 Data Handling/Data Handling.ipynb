# Exercise

```python
import csv

filtered_criteria = ['A','B']
filtered_rows = []

with open('students.csv','r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        if row['Grade'] in filtered_criteria:
            row['Status'] = 'Passed'
            filtered_rows.append([row['Name'],row['Age'],row['Grade'],row['Status']])
            
with open('top_students.csv','w',newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['Name','Age','Grade','Status'])
    writer.writerows(filtered_rows)
```

# Practice exercise

```python
import requests

api_url = "https://jian.sh/malaysia-api/state/v1/all.json"
response = requests.get(api_url)
malaysia_state_data = response.json()

while True:
    state_name = input("Enter the name of a state: ")

    state_found = False
    for state in malaysia_state_data:
         if state_name.lower() == state['state'].lower():
            print(f"Population of {state_name}: {state['population']}")
            state_found = True
            break

    if not state_found:
        print(f"{state_name} is not found in the data")
```
